```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sklearn.datasets import load_boston
import warnings
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def mean_abs_percent(data, model, real_val):
    y_pred = model.predict(data).reshape(-1)
    return np.mean(np.abs(y_pred - real_val)/real_val)*100
```

```python
with warnings.catch_warnings():
    # You should probably not use this dataset.
    warnings.filterwarnings("ignore")
    X, y = load_boston(return_X_y=True)
print(X.shape, y.shape)
```

```python
fig, axs = plt.subplots(2, figsize=(17,20))

fig.suptitle('нейросети с loss = MeanAbsoluteError')
axs[0].set_title('Loss')
axs[1].set_title('Metric mean_absolute_percentage_error')

# число рабочих эпох
epochs = 20
# вектор для выводв результатов
epoch = np.arange(epochs)
label = {}
for i_layers in [50, 100, 200]:
    for i_activation in ['relu', 'tanh', 'elu']:
        for i_optim in ['RMSProp','adam','NAdam']:
            # создаем рабочую модель modeli куда будем загружать эти вес (она должна быть идентичной структуры)
            modeli = Sequential([
                                Dense(128, activation='relu', input_shape=(13,)),
                                Dense(i_layers+10, activation=i_activation),
                                Dense(i_layers-10, activation=i_activation),
                                Dense(1, activation='linear'),
                                ])
            # компилируем modeli с одним из оптимизаторов
            modeli.compile(
                optimizer=i_optim,
                loss= 'mean_squared_logarithmic_error',#tf.keras.losses.MeanAbsoluteError(),
                metrics=['mean_absolute_percentage_error'],
                )
            # вычисляем ошибку для modeli без обучения
            hh0_tr = modeli.evaluate(X[:400,:], y[:400], verbose = 0)
            # проводим обучение модели 
            hh = modeli.fit(
                            X[:400,:], 
                            y[:400],
                            epochs=epochs,
                            batch_size=32, validation_split=0.2
                            )
            test_values = mean_abs_percent(X[400:,:], modeli, y[400:])
            name_label = f'h_layers={i_layers}, activ={i_activation}, optim={i_optim}'
            label.update({name_label:test_values})
            print(name_label, ' -- ', test_values)
            axs[0].plot(epoch, hh.history['loss'],'-',label = name_label+' train')
            axs[1].plot(epoch, hh.history['mean_absolute_percentage_error'],'-',label = name_label+' train')
            axs[1].plot(epoch,[test_values for i in range(len(epoch))],'--',label = name_label+' test')
            
axs[0].set_yscale('log')
axs[1].set_yscale('log')
axs[0].legend()
axs[1].legend()
axs[0].grid('On')
axs[1].grid('On')

axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Mean squared logarithmic error')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Mean abs percentage error')
plt.show()
```

## mean_squared_logarithmic_error loss показала себя лучше многих других метрик в среднем для различных комбинаций оптимизаторов, количества нейронов с слоях и функций активации с параметрами оптимизаторов по умолчанию
Нельзя однозначно сказать что меньшее или большее число нейронов скрытых слоев ведет к улучшению метрики качества.

```python
print("{:^45} {:^25}".format('Net parameters','Metric (mean abs % error)'))
for k, v in label.items():
    print("{:^45} {:^25}".format(k, round(v, 2)))
```
